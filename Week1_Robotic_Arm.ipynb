```python
import math

# Target Position
x = 8
y = 6

# Link Lengths
L1 = 5
L2 = 5

# Inverse Kinematics
cos_theta2 = (x**2 + y**2 - L1**2 - L2**2)/(2*L1*L2)

theta2 = math.acos(cos_theta2)

theta1 = math.atan2(y, x) - math.atan2(
    L2*math.sin(theta2),
    L1 + L2*math.cos(theta2)
)

print("Target Position:")
print("X =", x)
print("Y =", y)

print("\nJoint Angles:")
print("Theta1 =", round(math.degrees(theta1),2))
print("Theta2 =", round(math.degrees(theta2),2))
```
